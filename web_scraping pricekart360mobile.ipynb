```python
price = []
title = []
display = []
cameraMAin = []
cameraBack = []
processor = []
ram = []
battery = []
dualCam =[]
```

```python
import requests
from bs4 import BeautifulSoup





url = "https://www.pricekart.com/mobile/apple-mobiles-price-list-in-india"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
    

items = soup.findAll('div',class_="col-sm-12 col-xs-12 l")

for item in items:
        price1 = item.find('div',class_="price").text
        title1 = item.find('div',class_="mkt-ttl245").text
        specs_ul = item.find('div',class_="col-sm-8 col-md-9 col-xs-8 mobile-left-info")
        specs_li = specs_ul.find_all('li')
        dis1 = specs_li[1].text
        camMain1 = specs_li[5].text
        camBack1 = specs_li[6].text
        dualcam1 = specs_li[7]
        processor1 = specs_li[8].text
        ram1 = specs_li[9].text
        bettery1 = specs_li[10].text
    
        price.append(price1)
        title.append(title1)
        display.append(dis1)
        cameraMAin.append(camMain1)
        cameraBack.append(camBack1)
        dualCam.append(dualcam1)
        processor.append(processor1)
        ram.append(ram1)
        battery.append(bettery1)
```

```python
print(price)
print(title)
print(display)
print(cameraMAin)
print(cameraBack)
print(processor)
print(ram)
print(battery)
##print(dualCam)
```

```python
df1=pd.DataFrame({'Product Price':price,'Product Name':title,'Ram':ram,'Display':display,'Cameras Main':cameraMAin,'Cameras Back':cameraBack,'Processor':processor,'Battery':battery})
```
